```
#!/usr/bin/env python
# coding: utf-8

# # Линеарна регресија
# ## Вовед
# Повеќето статистички испитувања се однесуваат на повеќе карактеристики за дадена популација.  За да се испита врската помеѓу две или повеќе карактеристики, потребно е тие карактеристики да се мерат на иста група единки. Природата и јачината на врската помеѓу две или повеќе квантитативни карактерстики може да се испита со две познати техники во статистиката, регресиона и корелациона анализа, кои иако се поврзани имаат различни намени. Регресионата анализа се користи за одредување на видот на врската (функционалната поврзаност) на карактеристиките и главната цел кога се користи овој метод е да се предвиди или процени вредноста на едната променлива за дадена вредност на другата променлива.
# 
# Корелационата анализа, од друга страна, е поврзана со мерење на јачината на врската помеѓу променливите, односно го определува степенот на корелација помеѓу променливите. Честопати во пракса се сретнуваме со мерења на повеќе карактеристики на исти единки од популацијата. Вакви мерења најчесто се прават за да се провери како промените на одредени карактеристики (независни променливи) влијаат на промените на карактеристиката од интерес (зависна променлива).
# 
# Од аспект на машинското учење, надгледувано учење е пристап каде алгоритамот учи од дадени податоци за тренирање за да прави предвидувања каде е даден "точниот одговор" за секој примерок од податоците (излезната променлива). Надгледуваното учење ги вклучува класификацијата и регресијата.
# 
# Линеарната регресија е статистички метод, а и специфичен алгоритам во рамките на надгледуваното учење, кој се користи за моделирање на врската помеѓу зависна променлива и една или повеќе независни променливи (карактеристики). Целта на линеарната регресија е да се најде линија која најдобро одговара да ја опише врската помеѓу зависната променлива и независните променливи. Линеарната регресија претпоставува линеарност и регресионата линија се обидува да го минимизира збирот на квадратните разлики помеѓу вистинските вредности и предвидените вредности.
# 
# Ова укажува дека, надгледуваното учење е поширок концепт кој опфаќа различни алгоритми, додека линеарната регресија е специфичен алгоритам што се користи за проблеми со регресија со претпоставка за линеарна врска. Линеарната регресија е еден од најчесто користените методи за анализа на податоци и широко се користи во области како што се финансиите, економијата и науката.
# 
# Постојат неколку претпоставки кои мора да се исполнат за да биде валидна линеарната регресија: 1) линеарност: врската помеѓу зависната променлива и независните променливи мора да е линеарна; 2) независност: набљудувањата се независни едни од други; 3) хомогеност на варијансата: варијансата на грешките е константна на сите нивоа на независните променливи; 4) нормалност: грешките се нормално распределени; 5) без мултиколинеарност: не постои совршена корелација помеѓу кои било две независни променливи.
# 
# Равенката на регресионата линија е претставена со:
# 
# \begin{equation}
# y = f(x_1,x_2, ... , x_n) + \epsilon
# \end{equation}
# 
# или со
# 
# \begin{equation}
# y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n + \epsilon
# \end{equation}
# 
# каде што $y$ е зависната променлива, $\beta_0$ е пресекот (intercept), $\beta_1$, $\beta_2$, ... , $\beta_n$ се коефициентите на независните променливи $x_1$, $x_2$, ..., $x_n$, а $\epsilon$ е член за грешка.
# 
# Коефициентите $\beta_1$, $\beta_2$, ..., $\beta_n$ ја претставуваат промената на зависната променлива $y$ кога независните променливи $x_1$, $x_2$, ..., $x_n$ се менуваат за една единица. Пресекот $\beta_0$ ја претставува вредноста на зависната променлива кога сите независни променливи се еднакви на нула.
# 
# Постојат два вида линеарна регресија: едноставна линеарна регресија и повеќекратна линеарна регресија. Во едноставна линеарна регресија, постои само една независна променлива, додека во повеќекратна линеарна регресија, постојат две или повеќе независни променливи.
# 
# ## Едноставна линеарна регресија
# 
# Едноставна линеарна регресија се користи за моделирање на врската помеѓу зависна променлива и една независна променлива. Равенката на регресионата линија во едноставна линеарна регресија е претставена како:
# 
# \begin{equation}
# y = \beta_0 + \beta_1 x + \epsilon
# \end{equation}
# 
# каде што $y$ е зависната променлива, $x$ е независна променлива, $\beta_0$ е пресекот, $\beta_1$ е коефициентот на независната променлива $x$, а $\epsilon$ е член на грешка.
# 
# Коефициентот $\beta_1$ ја претставува промената на зависната променлива $y$ кога независната променлива $x$ се менува за една единица. Пресекот $\beta_0$ ја претставува вредноста на зависната променлива кога независната променлива е еднаква на нула.
# 
# За да се најдат вредностите на $\beta_0$ и $\beta_1$, може да се користи методот на најмали квадрати. Овој метод вклучува пронаоѓање на вредностите на $\beta_0$ и $\beta_1$ кои го минимизираат збирот на квадратните разлики помеѓу вистинските вредности и предвидените вредности.
# 
# ## Повеќекратна линеарна регресија
# 
# Повеќекратна линеарна регресија се користи за моделирање на врската помеѓу зависна променлива и две или повеќе независни променливи. Равенката на регресиoната линија во повеќекратна линеарна регресија е претставена како:
# 
# \begin{equation}
# y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n + \epsilon
# \end{equation}
# 
# каде $y$ е зависната променлива, $x_1$, $x_2$, ..., $x_n$ се независните променливи, $\beta_0$ е пресекот, $\beta_1$, $\beta_2$, ..., $\beta_n$ се коефициентите на независните променливи $x_1$, $x_2$, ..., $x_n$, а $\epsilon$ е член на грешка.
# 
# За да ги најдеме вредностите на $\beta_0$, $\beta_1$, $\beta_2$, ..., $\beta_n$ може да се користи методот на најмали квадрати. Овој метод вклучува наоѓање на вредностите на $\beta_0$, $\beta_1$, $\beta_2$, ..., $\beta_n$ кои го минимизираат збирот на квадратните разлики помеѓу вистинските вредности и предвидените вредности.
# 
# ## Полиномно проширување на линеарниот модел
# Во линеарна регресија, моделот претпоставува линеарна врска помеѓу влезните карактеристики и излезната променлива. Меѓутоа, во некои случаи, врската може да не е строго линеарна, а вклучувањето на полиномните членови може да помогне да се доловат нелинеарни обрасци во податоците. Проширувањето на линеарната регресија во полиномен модел вклучува додавање полиномни карактеристики на влезните податоци.
# 
# За да се прошири линеарната регресија во полиномен модел, се воведуваат дополнителни карактеристики кои се степени на оригиналните карактеристики. На пример, ако имаме единечна влезна променлива $x$, можеме да се прошири во полиномен модел од степен 2 со вклучување на карактеристиката $x^2$ како дополнителна влезна карактеристика. Ова му овозможува на моделот да фати квадратни односи помеѓу карактеристиката и целната променлива.
# 
# Општата равенка за модел на полиномна регресија со степен $n$ е:
# \begin{equation}
# y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + ... + \beta_n x^n + \epsilon
# \end{equation}
# 
# каде што $y$ е излезната променлива, $x$ е влезната карактеристика, $\beta_0$, $\beta_1$, $\beta_2$, ..., $\beta_n$ се коефициентите, $x^2$, $x^3$, ..., $x^n$ се полиномните карактеристики и $\epsilon$ е членот за грешка.
# 
# 
# ## Справување со мултиколинеарност, повеќедимензионалност и префитување
# Мултиколинеарноста се јавува кога две или повеќе независни променливи во повеќекратен линеарен регресивен модел се во голема корелација една со друга. Ова може да доведе до неверодостојни проценки на коефициентите и намалување на моќта на предвидување на моделот. За да се справиме со мултиколинеарноста, може да се: отфрли една од високо корелираните независни променливи, да се користи анализа на главна компонента (Principal component analysis - PCA) за да се намали димензионалноста на множеството или пак, да се користат техники за регуларизација како ласо (lasso) регресија, риџ (ridge) регресија и регресија на еластична мрежа (elastic net).
# 
# ## Ласо (Lasso) регресија
# Ласо регресијата е само една од многуте техники за регуларизација што може да се користат за да се спречи префитување во моделите на линеарна регресија. Ласо регресијата е техника на линеарна регресија која истовремено врши и селекција на карактеристики и регуларизација. Тоа е корисно во ситуации каде што има многу карактеристики или влезни променливи, од кои некои може да бидат ирелевантни или да имаат мал ефект врз променливата за одговор. Во такви случаи, ласо регресијата може да се користи за да се намали бројот на карактеристики што се користат во моделот, што резултира со поедноставен и поинтерпретабилен модел.
# 
# Клучната идеја зад ласо регресијата е да се додаде $L_1$ казна на функцијата на трошоците на линеарната регресија. $L_1$ казната е едноставно збир на апсолутните вредности на коефициентите на карактеристиките. Овaa казнa принудува некои од коефициентите да бидат нула, со што ефективно врши селекција на карактеристики. Степенот на реткост на добиениот модел може да се контролира со хиперпараметар наречен $\alpha$.
# 
# Функцијата на трошоците на ласо регресијата е дадена со:
# 
# \begin{equation}
# J(\beta)=1/2m * \sum_{i=1}^{m}[y_i-\beta_0-\sum_{j=1}^{n}(\beta_j x_{ij})]^2 + \alpha \sum_{j=1}^{n} |\beta_j|
# \end{equation}
# 
# каде што $m$ е бројот на примероци, $n$ е бројот на карактеристики, $y_i$ е излезната променлива на $i$-тиот примерок, $x_{ij}$ е вредноста на $j$-та карактеристика за $i$-тиот примерок, $\beta_j$ е коефициентот на $j$-тата карактеристика, $\beta_0$ е пресекот и $\alpha$ е параметарот за регуларизација.
# 
# За да се реши проблемот со оптимизацијата, може да се користи алгоритам за спуштање на градиентот (gradient descent), кој вклучува ажурирање на секој коефициент додека другите се фиксирани.
# 
# ## Риџ (Ridge) регресија
# Риџ регресијата е техника на линеарна регресија која врши регуларизација со додавање на член на $L_2$ казна на функцијата на трошоци. Оваа техника е корисна во ситуации каде што има многу карактеристики или влезни променливи, од кои некои може да бидат во корелација или да имаат мали ефекти врз излезната променлива. Риџ регресијата може да се користи за да се намали варијансата на моделот и да се подобрат неговите перформанси на генерализација.
# 
# Клучната идеја зад риџ регресијата е да се додаде $L_2$ казнана функцијата на трошоците на линеарната регресија. $L_2$ казната е едноставно збир на квадратите на коефициентите на карактеристиките. Оваа казна ги принудува коефициентите да бидат мали, со што ефективно врши регуларизација. Степенот на регуларизација може да се контролира со хиперпараметар наречен $\alpha$.
# 
# Функцијата на трошоците на риџ регресијата е дадена со:
# 
# \begin{equation}
# J(\beta)=1/2m * \sum_{i=1}^{m}[y_i-\beta_0-\sum_{j=1}^{n}(\beta_j x_{ij})]^2 + \alpha \sum_{j=1}^{n} \beta_j^2
# \end{equation}
# 
# каде што $m$ е бројот на примероци, $n$ е бројот на карактеристики, $y_i$ е излезната променлива на $i$-тиот примерок, $x_{ij}$ е вредноста на $j$-та карактеристика за $i$-тиот примерок, $\beta_j$ е коефициентот на $j$-тата карактеристика, $\beta_0$ е пресекот и $\alpha$ е параметарот за регуларизација.
# 
# За да се реши проблемот со оптимизацијата, може да се користи решение во затворена форма, кое вклучува земање на изводот на функцијата на трошоците во однос на коефициентите и негово поставување на нула. Ова резултира со едноставен линеарен систем на равенки што може ефикасно да се реши со помош на линеарна алгебра.
# 
# ## Регресија на еластична мрежа (Elastic net)
# Еластична мрежа е техника на линеарна регресија која ги комбинира $L_1$ и $L_2$ казните на ласо и риџ регресија, соодветно. Оваа техника е корисна во ситуации каде што има многу карактеристики или влезни променливи, од кои некои може да бидат во корелација или да имаат мали ефекти врз излезната променлива. Регресијата на еластична мрежа, како и претходната, исто така, може да се користи за да се намали варијансата на моделот и да се подобри неговата изведба на генерализација.
# 
# Клучната идеја зад регресијата на еластична мрежа е да се додаде казна што ги комбинира $L_1$ и $L_2$ казните на функцијата на трошоците на линеарната регресија. $L_1$ казната е збир на апсолутните вредности на коефициентите на карактеристиките, додека $L_2$ казната е збирот на квадратите на коефициентите на карактеристиките. Казната се контролира со два хиперпараметра: $\alpha$ и $l1_{ratio}$. $\alpha$ параметарот го контролира степенот на регуларизација, додека параметарот $l1_{ratio}$ ја контролира рамнотежата помеѓу $L_1$ и $L_2$ казните.
# 
# Функцијата на трошоците на регресијата на еластична мрежа е дадена со:
# 
# \begin{equation}
# J(\beta)=1/2m * \sum_{i=1}^{m}[y_i-\beta_0-\sum_{j=1}^{n}(\beta_j x_{ij})]^2 + \alpha l1_{ratio} \sum_{j=1}^{n} |\beta_j| + \alpha(1-l1_{ratio})/2 \sum_{j=1}^{n} \beta_j^2
# \end{equation}
# 
# каде што $m$ е бројот на примероци, $n$ е бројот на карактеристики, $y_i$ е излезната променлива на $i$-тиот примерок, $x_{ij}$ е вредноста на $j$-та карактеристика за $i$-тиот примерок, $\beta_j$ е коефициентот на $j$-тата карактеристика, $\beta_0$ е пресекот и $\alpha$ е параметарот за регуларизација и $l1_{ratio}$ е односот на $L_1$ и $L_2$ казните.
# 
# Изборот на техниката на регуларизација и вредноста на хиперпараметарот(ите) зависат од податочното множество и конкретниот проблем што се разгледува.
# 
# ## Баесова риџ регресија
# Баесовата риџ регресија е линеарен регресивен алгоритам кој вклучува баесов пристап за проценка на параметрите на моделот. Претпоставува претходна распределба над параметрите на моделот и ја ажурира оваа распределба врз основа на набљудуваните податоци, што резултира со постериорни распределби за параметрите. Баесовата рамка обезбедува начин за квантифицирање на несигурноста во проценките на параметрите.
# 
# ## Евалуација на перформансите на моделот
# 
# Со цел да се процени перформансите на повеќекратниот линеарен регресивен модел, постојат различни статистички метрики кои можат да се користат. Некои од нив се:
# 
# * R-квадрат: Ова е мерка за тоа колку од варијансата во зависната променлива се објаснува со независните променливи во моделот. Повисоката вредност на R-квадрат покажува подобро вклопување на моделот.
# 
# * Средна квадратна грешка (Mean squared error - MSE): Ова е просечната квадратна разлика помеѓу реалните и предвидените вредности на зависната променлива. Пониска вредност на MSE укажува на подобро вклопување на моделот.
# 
# * Средна апсолутна грешка (Mean absolute error - MAE): За разлика од средната квадратна грешка (MSE) која ја пресметува просечната квадратна разлика помеѓу предвидените и вистинските вредности, MAE ја мери просечната апсолутна разлика.
# 
# * Корен на средна квадратна грешка (Root mean squared error - RMSE): Ова е квадратниот корен на MSE. Тоа е мерка за просечното отстапување на предвидените вредности од вистинските вредности.

# ## Примери за линеарна регресија со различни податочни множества: Едноставна линеарна регресија со податочно множество за оценки на студенти
# 
# Податочното множество за оценки на студенти, со кое ќе прикажеме едноставна линеарна регресија, се состои од 25 податоци за студенти. За секој секој студент се дадени часовите кои ги потрошиле за учење на испит во влезната променлива (карактеристика, атрибут) "Hours" и поените кои ги добиле како оценка по полагањето на испитот во излезната променлива "Scores".
# 
# Во оваа секција најпрво ќе покажеме како може да се вчита податочното множество, а потоа ќе покажеме како може да се направи основна анализа и визуелизација на податоците, за да на крај ја прикажеме линеарната зависност помеѓу часовите за учење и поените добиени на испитот, со помош на едноставен модел на линеарна регресија.
# 
# ### Вчитување на потребните библиотеки
# Најпрво треба да ги вчитаме библиотеките кои се потребни за извршување на кодот во оваа задача.

# In[91]:


import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt # Библиотека за визуелизација
import seaborn as sns
import numpy as np # Библиотека за нумерички анализи
import pandas as pd
from scipy import stats # Библиотека за статистички анализи
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler # Min max Scaler за нормализација на податоци во опсег [0,1]
from sklearn.linear_model import LinearRegression  # Вчитување на модели за линерна регресија
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score # За евалуација на модели


# ### Вчитување на податоците
# Податочното множество може да го вчитаме со функциски повик за читање на .csv датотеки како "pandas" податочна рамка.

# In[92]:


data = pd.read_csv("datasets/student_scores.csv") # Вчитување на податочното множество


# Со head(n) може да ги прикажеме првите n примероци.

# In[93]:


data.head(10)


# ### Визуeлизација на податоците
# За да ја увидиме линерната врска помеѓу влезната променлива за часовите на учење на испитот и излезната променлива за поените добиени како оценка по полагањето на испитот, може да ги визуелизираме податоците со дијаграм на растурање. Од дијаграмот се гледа дека постои некоја "привидна" линеарна зависност меѓу влезната и излезната променлива.

# In[94]:


# Приказ на податоците со дијаграм на растурање
plt.figure(figsize=(5, 4))
plt.scatter(data["Hours"], data["Scores"], color = 'darkgreen')
plt.xlabel("Часови на учење")
plt.ylabel("Оцена на испитот")
plt.show()


# Ќе ги вчитаме часовите за учење на испитот во влезната променлива X и добиената оценка во излезната променлива y. Поставуваме дека има само еден атрибут (карактеристика) која ги опишува податоците за часовите за учење и еден излез за оценката. Распределбата на вредностите за атрибутите може да ја разгледаме со користење на хистограм со помош на функцијата sns.histplot(). Тука имаме само еден атрибут, па затоа имаме еден хистограм, на кој со опцијата kde=True прикажуваме и апроксимативна крива која ја отсликува распределбата на вредностите.
# Истото го правиме за да прикажеме хистограм на вредностите на излезната променлива за добиената оценка на испитот. Може да забележиме дека обликот на хистограмите е сличен, освен за почетните вредности, што укажува на можност за постоење линеарна врска меѓу влезната и излезната променлива.

# In[95]:


number_of_features = 1
number_of_targets = 1
X = data["Hours"].values.reshape(-1,1)
y = data["Scores"].values
N = len(y)

fig = plt.figure(figsize=(4, 4))
fig.subplots(nrows=1, ncols=1)
for feat_i in range(number_of_features):
    ax = plt.subplot(1,1, feat_i+1)
    sns.histplot(X[:,feat_i],color='navy',label="сите класи",kde=True,stat="count") # се прикажува хистограм на вредностите на соодветниот атрибут без разлика во која класа припаѓаат
    plt.legend()
plt.show()

fig = plt.figure(figsize=(4, 4))
sns.histplot(y, color = "navy")
plt.show()


# ### Линеарна регресија
# За да се вклопи (фитува) линеарен регресивен модел на податоците, ќе создадеме објект LinearRegression и ќе го користиме методот fit() за да го обучиме моделот.

# In[96]:


reg = LinearRegression()
reg.fit(X, y)


# Сега, можеме да го користиме моделот за да правиме предвидувања за нови податоци.

# In[97]:


new_data = pd.DataFrame([7, 8, 9])
predictions = reg.predict(new_data)
print(predictions)


# ### Нормализација на податоците
# Ќе направиме мин-макс нормализација на податоците со MinMaxScaler() објект. Со fit_transform(X) ги нормализираме влезните податоци со помош на MinMaxScaler, во опсегот помеѓу 0 и 1. Потоа, ги делиме нормализираните податоци во две групи - тренирачка група и тест група.
# X_train содржи тренирачките влезни податоци (обележени со X) кои се користат за обучување на моделот, X_test содржи тест влезни податоци (обележени со X) кои се користат за проверка на моделот, y_train содржи тренирачки излезни податоци (обележени со y) кои се користат за обучување на моделот, a y_test содржи тест излезни податоци (обележени со y) кои се користат за проверка на моделот. Со test_size=0.30 го поставуваме односот помеѓу тренирачките и тест податоци, 30% за тест. random_state ја поставува семенската вредност (seed value) на случајната генерација што овозможува повторливост на поделбата на податоците при секое извршување на кодот.

# In[98]:


scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
#Забележете дека бидејќи ова е малo податочно множество, не треба да го користиме стандардниот train_test_split
#за да ја процениме точноста на множеството податоци, туку треба да користиме вкрстена валидација.
#Меѓутоа, за едноставност, го правиме тоа во овој пример.


# ### Пресметка на грешки
# Потоа, правиме линеарна регресија врз тренирачките податоци X_train и соодветните излезни вредности y_train. Потоа, се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Со помош на reg.predict(X_test), се предвидуваат вредностите на излезните податоци y за тест податоците X_test. За оценување на фитуваниот модел на линерна регресија пресметуваме различни метрики: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и
# R-квадрат. Може да забележиме дека грешките не се големи, а R-квадрат вредноста оди кон 1, што укажува на тоа дека моделот е подобар во објаснувањето на варијациите во податоците.

# In[99]:


reg = LinearRegression().fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred = reg.predict(X_test) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test,y_pred)) # пресметај ја R - квадрат вредноста


# Потоа може да направиме визуелизација на предвидените вредности од моделот во однос на вистинските вредности со помош на дијаграм на растурање, од каде што повторно може да се забележи линеарната врска во овој пример.

# In[100]:


# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# ### Приказ на линијата која најдобро одговара на моделот
# За да ја прикажеме линијата која најдобро одговара на линеарната регресија, може да иницијализираме повторно модел на линеарна регресија и да го поставиме на податоците. Ја пресметуваме варијансата на моделот и креираме опсег на влезните вредности line_X за претставување на линијата на моделот, а потоа предвидуваме излезни вредности line_y со користење на моделот. Со помош на дијаграм на растурање може да ги
# прикажеме податочните точки, со plt.plot() линијата која најдобро одговара на моделот, а со plt.errorbar() грешката на моделот.

# In[101]:


# поставете модел на линеарна регресија
lr = LinearRegression() # иницијализација на моделот
lr.fit(X, y) # поставете го моделот на податоците
lrsigma = np.sum((y - lr.predict(X))**2) /(N-2) # пресметка на варијансата на моделот
# пресметка на линијата на модел во опсег [20, X.max()]
line_X = np.arange(0, 1.1, 0.1)[:, np.newaxis]
line_y = lr.predict(line_X)

lw = 2 # параметар за визуелизација (linewidth)
plt.figure(figsize=(5, 4)) # ги цртаме податочните точки и ги цртаме линиите што го претставуваат моделот, ги цртаме и лентите за грешки на моделот за линеарна регресија
plt.scatter(X, y, color='darkgreen', marker='.', s = 300)
plt.plot(line_X, line_y, color='navy', linewidth=lw, label='Линеарна регресија', alpha=0.5)
plt.errorbar(line_X, line_y,yerr=lrsigma)

plt.legend(loc='best')
plt.xlabel("Влез: мин-макс нормализирани часови на учење")
plt.ylabel("Излез: Оцена на испитот")
plt.show()


# ## Повеќекратна линеарна регресија со California Housing податочното множество
# 
# Податочното множество Califonia Housing се состои од 20 640 примероци од сместувања во различни делови на Калифорнија. За секое сместување се дадени вредностите на 8 атрибути: "MedInc" (медијана на примањата), "HouseAge" (старост на сместувањето), "AveRooms" (среден број на соби),	"AveBedrms" (среден број на спални соби), "Population" (популација), "AveOccup" (средна пополнетост), "Latitude" (географска ширина), "Longitude" (географска должина), "Median House Value" (средна вредност на сместувањето). Излезната променлива е "MedHouseVal" која ја претставува средната вредност како цена на сместувањето. Целта е да поставиме модел кој најдобро одговара на повеќекратна регресија која ќе ја предвиди оваа излезна цена со податоците од влезните 8 атрибути. Ќе ја испитаме зависноста на секој атрибут со излезната променлива, за да може да се најдат атрибутите кои најмногу може да придонесат за подобар регресионен модел.
# 
# Прво, ќе покажеме како може да се вчита податочното множество, а потоа како може да се направи основна анализа и визуелизација на податоците, за да на крај ја прикажеме можноста за вклопување на различни модели на регресија.
# 
# #### Вчитување на потребните библиотеки
# Најпрво треба да ги вчитаме библиотеките кои се потребни за извршување на кодот во оваа задача.

# In[102]:


import matplotlib.pyplot as plt #Библиотека за визуелизација
import seaborn as sns
import numpy as np #Библиотека за нумерички анализи
import pandas as pd
from scipy import stats #Библиотека за статистички анализи
from sklearn.model_selection import train_test_split
from sklearn import datasets #Библиотека за податочни множества за машинско учење
from sklearn.preprocessing import MinMaxScaler #Min max Scaler за нормализација на податоци во опсег [0,1]
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge #Вчитување на модели за линерна регресија
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score #За евалуација на модели
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.datasets import fetch_california_housing


# ### Вчитување на податоците
# Податочното множество може да го вчитаме со функциски повик за читање од библиотеката sklearn.datasets.

# In[103]:


california_housing = datasets.fetch_california_housing()
df = datasets.fetch_california_housing(as_frame=True).frame


# ### Визуелизација на податоците
# Ќе ги прикажеме првите 10 податоци од множеството.

# In[104]:


df.head(10)


# Ја иницијализираме влезната променлива X, каде што секој ред одговара на податочна точка, а секоја колона одговара на карактеристика (атрибут), a и излезната променлива y каде што i-тата вредност е излезната цена на i-тиот вектор на атрибут (ред) во X. Потоа, може да го испечатиме името на излезната променлива ("MedHouseVal"), бројот на атрибути (8), бројот на излезни променливи (1) и бројот на инстанци (20640) во множеството.

# In[105]:


X = california_housing.data # Х матрицата одговара на влезните вектори, каде што секој ред одговара на податочна точка, а секоја колона одговара на карактеристика (атрибут)
y = california_housing.target # Векторот y одговара на излезните вредности, каде што i-тата вредност е класата на i-тиот вектор на атрибут (ред) во X
number_of_features=X.shape[1]
number_of_targets=1
feature_names = california_housing.feature_names
target_names = california_housing.target_names
print("Имиња на излезните класи: ", target_names)
N = X.shape[0]

print("Број на карактеристики (атрибути): ", X.shape[1]) #бројот на атрибути одговара на бројот на колони во X матрицата
print("Број на излезни класи: ", len(target_names)) # или number_of_targets
print("Број на инстанци: ", X.shape[0]) # или N


# Распределбата на вредностите за атрибутите може да ја разгледаме со користење на хистограм. Тука имаме само 8 атрибути, па затоа имаме 8 хистограми посебно, на секои од нив прикажуваме и апроксимативна крива која ја отсликува распределбата на вредностите.

# In[106]:


fig = plt.figure(figsize=(8, 15))
fig.subplots(nrows=4, ncols=2)
for feat_i in range(number_of_features): #хистограми за секој атрибут
    ax = plt.subplot(4,2, feat_i+1)
    plt.title(feature_names[feat_i])
    sns.histplot(X[:,feat_i],color='navy',label='сите класи',kde=True,stat="count") # се прикажува хистограм на вредностите на соодветниот атрибут без разлика во која класа припаѓаат
    plt.legend()
plt.show()


# Истото го правиме за да прикажеме хистограм на вредностите на излезната променлива. Може да забележиме дека обликот на хистограмите на влезните атрибути, а и на излезната променлива, не е многу сличен, што може да укаже на различни распределби за секоја променлива, а и не голема линеарна зависност помеѓу секој влезен атрибут со излезната променлива.

# In[107]:


fig = plt.figure(figsize=(5, 5))
sns.histplot(y, color = "navy")
plt.show()


# ### Визуелизација на зависности
# Може да ја прикажеме зависноста на секој атрибут со секој друг атрибут со визуелизација на атрибутите (карактеристиките) преку дијаграми на растурање. Прикажуваме 8x8=64 вакви дијаграми, по еден за секоја комбинација на два атрибути. Оваа визуелизација ни помага да го истражиме односот помеѓу различните атрибути во податочното множество и да забележиме можни зависности или корелации помеѓу нив. На прв поглед, дијаграмите се многу "растурени", а визуелно може да се забележи само некоја мала линеарна зависност меѓу атрибут 6 и 7.

# In[108]:


fig = plt.figure(figsize=(15, 15))
fig.subplots(nrows=number_of_features, ncols=number_of_features)
for feature_i in range(number_of_features):
    for feature_j in range(number_of_features):
        ax = plt.subplot(number_of_features,number_of_features,number_of_features*feature_i + feature_j+1)
        if feature_i==feature_j:
          plt.hist(X[:, feature_i], alpha=.6, color="orange")
          plt.title("Ат. "+str(feature_i)+" (хистограм)")
        else:
          plt.scatter(X[:,feature_i], X[:,feature_j], color="darkorange")
          plt.title("Ат. "+str(feature_i)+" x Ат. "+str(feature_j)) # Атрибут i x Атрибут ј
          #plt.title(feature_names[feature_i] +" x "+feature_names[feature_j]) # Атрибут i x Атрибут ј
        plt.xticks(())
        plt.yticks(())
plt.show()


# Може да ја прикажеме зависноста на секој атрибут и со излезната променлива преку дијаграми на растурање. Прикажуваме 8 вакви дијаграми, по еден за секој атрибут со излезната променлива. Оваа визуелизација ни помага да го истражиме односот помеѓу различните атрибути во податочното множество и излезната променлива и да забележиме кој атрибут би можело да има линеарна зависност со излезот.

# In[109]:


fig = plt.figure(figsize=(8, 15))
fig.subplots(nrows=(int)(number_of_features/2),ncols=2)
for feat_i in range(number_of_features):
    ax = plt.subplot(4,2, feat_i+1)
    plt.scatter( X[:,feat_i], y, color="darkorange")
    plt.xticks(())
    plt.yticks(())
    plt.title("Атрибут "+str(feat_i)+" x Излез")
plt.show()


# На претходните дијаграми може визуелно да се увиди кои атрибути се поврзани помеѓу себе. Сепак, за тоа може да ги изразиме зависностите квантитативно, со помош на Пирсоновиот коефициент на корелација. Затоа, пресметуваме и визуелизираме корелациска матрица во која се претставени Пирсоновите коефициенти на корелација за сите парови атрибути, како и за секој атрибут со целната променлива, во последната редица и колона на матрицата. Од овде, може да се види зависноста меѓу атрибут 6 и 7 која ја споменавме, а и малата зависност на сите влезни променливи со излезната.

# In[110]:


correlation_matrix = np.zeros((number_of_features+1,number_of_features+1))
# прво ги пресметуваме присоновите корелациски коефициенти за сите парови атрибути
for feature_i in range(number_of_features):
    for feature_j in range(number_of_features):
        corr, _ = stats.pearsonr(X[:,feature_i], X[:,feature_j])
        correlation_matrix[feature_i][feature_j] = corr
# потоа ги пресметуваме пирсоновите корелациски коефициенти за секој атрибут со целната променлива
for feature_i in range(number_of_features):
    target_i = number_of_features
    corr, _ = stats.pearsonr(X[:,feature_i], y)
    correlation_matrix[feature_i][target_i] = corr
    correlation_matrix[target_i][feature_i] = corr
# на крај го пресметуваме пирсоновиот корелациски коефициент за целната променлива само со себе (ова може и да се изостави зашто резултатот е 1)
corr, _ = stats.pearsonr(y, y)
correlation_matrix[target_i][target_i] = corr
corr, _ = stats.pearsonr(X[:,feature_i], y)

plt.figure()
plt.imshow(correlation_matrix, cmap = "inferno", vmin=-1, vmax=1) # ја исцртуваме матрицата со imshow
plt.colorbar()
plt.show()


# Контурниот график на густина (jointplot од библиотеката seaborn со опцијата kind="kde") ни дава претстава за распределбата на податоците и можни области на концентрација на зависностите. Оваа визуелизација ни помага да го истражиме односот помеѓу атрибутите и излезните вредности и да забележиме можни зависности или корелации помеѓу нив. Ќе прикажеме еден ваков график за првиот атрибут со излезната променлива (се создава контурен график на густина помеѓу атрибутот X[:, 0] и излезните вредности y).

# In[111]:


sns.jointplot(x=X[:,0], y=y, kind="kde", space=0, color="g", height = 3)
plt.show()
# откоментирајте го следниот код за приказ за секој атрибут
#for feature_i in range(number_of_features):
#    sns.jointplot(x=X[:,feature_i], y=y, kind="kde", space=0, color="g", height = 3)
#plt.show()


# ### Нормализација на податоците
# Ќе направиме мин-макс нормализација на податоците со MinMaxScaler() објект, како и во претходниот пример за едноставна линеарна регресија. Потоа, ги делиме нормализираните податоци во две групи - тренирачка група и тест група. X_train содржи тренирачки влезни податоци (обележени со X) кои се користат за обучување на моделот, X_test содржи тест влезни податоци (обележени со X) кои се користат за проверка на моделот, y_train содржи тренирачки излезни податоци (обележени со y) кои се користат за обучување на моделот, a y_test содржи тест излезни податоци (обележени со y) кои се користат за проверка на моделот. Го поставуваме односот помеѓу тренирачките и тест податоци на 30% за тест.

# In[112]:


scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
#Забележете дека бидејќи ова е малo податочно множество, не треба да го користиме стандардниот train_test_split
#за да ја процениме точноста на множеството податоци, туку треба да користиме вкрстена валидација.
#Меѓутоа, за едноставност, го правиме тоа во овој пример.


# ### Линеарна регресија и пресметка на грешки
# Потоа, правиме повеќекратна линеарна регресија врз тренирачките податоци X_train и соодветните излезни вредности y_train. Потоа, се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Може да забележиме дека има 8 коефициенти, по еден за секој атрибут. Со помош на reg.predict(X_test), се предвидуваат вредностите на излезните податоци y за тест податоците X_test. За оценување на фитуваниот модел на линерна регресија пресметуваме различни метрики: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Може да забележиме дека грешките не се големи, но R-квадрат вредноста е околу 0.5, што укажува на тоа дека моделот не е многу добар во објаснувањето на варијациите во податоците.

# In[113]:


reg = LinearRegression().fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred = reg.predict(X_test) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test,y_pred)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# ### Риџ регресија
# Може да направиме риџ регресија врз тренирачките податоци X_train и соодветните излезни вредности y_train. Повикот Ridge(alpha=1).fit(X_train, y_train) креира инстанца на моделот риџ регресија со параметарот на регуларизација alpha=1 и го тренира моделот со користење на тренирачките податоци X_train и y_train. Потоа, се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Може да забележиме дека има 8 коефициенти, по еден за секој атрибут. Со помош на reg.predict(X_test), се предвидуваат вредностите на излезните податоци y за тест податоците X_test. За оценување на фитуваниот модел на риџ регресија пресметуваме различни метрики: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Може повторно да забележиме дека грешките не се големи, но R-квадрат вредноста е околу 0.5, што укажува на тоа дека моделот не е многу добар во објаснувањето на варијациите во податоците, иако има многу мало подобрување кога го направивме истово со линеарна регресија.

# In[114]:


reg = Ridge(alpha=1).fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred = reg.predict(X_test) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test,y_pred)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# ### Ласо регресија
# Исто така, може да направиме ласо регресија врз тренирачките податоци X_train и соодветните излезни вредности y_train. Повикот Lasso(alpha=0.1).fit(X_train, y_train) креира инстанца на моделот ласо регресија со параметарот на регуларизација alpha=0.1 и го тренира моделот со користење на тренирачките податоци X_train и y_train. Потоа, се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Може повторно да забележиме дека има 8 коефициенти, по еден за секој атрибут. Со помош на reg.predict(X_test), се предвидуваат вредностите на излезните податоци y за тест податоците X_test. За оценување на фитуваниот модел на ласо регресија ги пресметуваме метриките од претходно: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Може да забележиме дека грешките не се големи, но R-квадрат вредноста е намалена на 0.04, што укажува на тоа дека моделот не е добар во објаснувањето на варијациите во податоците.

# In[115]:


reg = Lasso(alpha = 0.1).fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred = reg.predict(X_test) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test,y_pred)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# ### Регресија на еластична мрежа
# Потоа, може да направиме регресија на еластична мрежа врз тренирачките податоци X_train и соодветните излезни вредности y_train. Повикот ElasticNet(alpha=0.1, l1_ratio=0.3).fit(X_train, y_train) креира инстанца на моделот еластилна мрежа со параметри alpha=0.1 и l1_ratio=0.3 и го тренира моделот со користење на тренирачките податоци X_train и y_train. Потоа, се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Може повторно да забележиме дека има 8 коефициенти, по еден за секој атрибут. Со помош на reg.predict(X_test), се предвидуваат вредностите на излезните податоци y за тест податоците X_test. За оценување на фитуваниот модел на еластична мрежа ги пресметуваме метриките: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Може да забележиме дека грешките не се големи, но R-квадрат вредноста е намалена од онаа во моделот на линеарна регресија и поголема од кај ласо регресијата. Моделот не е многу добар во објаснувањето на варијациите во податоците.

# In[116]:


reg = ElasticNet(alpha = 0.1, l1_ratio = 0.3).fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred = reg.predict(X_test) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test,y_pred)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# ### Баесова риџ регресија
# На крај, може да направиме баесова риџ регресија врз тренирачките податоци X_train и соодветните излезни вредности y_train. Повикот BayesianRidge().fit(X_train, y_train) креира инстанца на моделот Баесова риџ регресија и го тренира моделот со користење на тренирачките податоци X_train и y_train. Потоа, се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Може повторно да забележиме дека има 8 коефициенти, по еден за секој атрибут. Со помош на reg.predict(X_test, return_std = True), се предвидуваат вредностите на излезните податоци y за тест податоците X_test и нивната стандардна девијација y_std. За оценување на фитуваниот модел на регресија на еластична мрежа ги пресметуваме метриките: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Дополнително, може да ја испечатиме и стандардната девијација на предвидените излезни вредности. Може да забележиме дека грешките не се големи, но R-квадрат вредноста е слична како кај риџ регресијата. Моделот не е многу добар во објаснувањето на варијациите во податоците.

# In[117]:


reg = BayesianRidge().fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred, y_std = reg.predict(X_test, return_std = True) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print(y_std)
print(len(y_std))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test, y_pred)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# ### Споредба на моделите со приказ на линијата која најдобро одговара на податоците
# Ќе ги поставиме моделите на регресија (линеарна, риџ, ласо, еластична мрежа и Баесова риџ) повторно, но овој пат за да може да ги споредиме и тоа само да ја увидиме зависноста на првиот атрибут (атрибут 0) Xi во однос на излезната променлива y. Со користење на моделите, ги пресметуваме коефициентите за секој модел, кои ги опишуваат зависноста меѓу променливата Xi и излезната променлива y. Отпечатените коефициенти на моделите ни даваат информација за влијанието и значењето на Xi во предвидувањето на y. Што поголема вредност има коефициентот, тоа повеќе влијание има Xi врз y. Линеарната регресија ни дава коефициент на Xi кој е пропорционален со влијанието на Xi врз y. Риџ регресијата, ласо регресијата, регресијата на еластична мрежа и Баесова риџ регресијата користат различни регуларизациони техники за контролирање на комплексностa на моделот и различни тежини на коефициентите. Оваа анализа на коефициентите ни помага да го разбереме влијанието на конкретниот атрибут Xi врз излезната променлива y и да ги споредиме перформансите на различните модели на регресија во однос на овој атрибут.

# In[118]:


Xi=X[:,0].reshape(-1,1)
print("Xi=", Xi)
print("y=",y)

# поставете модел на линеарна регресија
lr = LinearRegression() # иницијализација на моделот
lr.fit(Xi, y) # поставете го моделот на податоците
lrsigma = np.sum((y - lr.predict(Xi))**2) /(N-2) # пресметка на варијансата на моделот

# поставете модел на риџ регресија
ridge = Ridge(alpha=1)
ridge.fit(Xi, y)
ridgesigma = np.sum((y - ridge.predict(Xi))**2) /(N-2)

# поставете модел на ласо регресија
lasso = Lasso(alpha=0.1)
lasso.fit(Xi, y)
lassosigma = np.sum((y - lasso.predict(Xi))**2) /(N-2)

# поставете модел на еластична мрежа
elastic = ElasticNet(alpha=0.1, l1_ratio = 0.4)
elastic.fit(Xi, y)
elasticsigma = np.sum((y - elastic.predict(Xi))**2) /(N-2)

# поставете модел на баесова риџ регресија
Bayes = BayesianRidge()
Bayes.fit(Xi, y)

print("Коефициенти на моделите")
print("Линеарна регресија: ",lr.coef_)
print("Риџ регресија: ", ridge.coef_)
print("Ласо регресија: ", lasso.coef_)
print("Регресија на еластична мрежа:", elastic.coef_)
print("Баесова риџ регресија: ", Bayes.coef_)


# За да ја прикажеме линијата која најдобро одговара на претходните модели, поставуваме линија line_X за претставување на линијата на моделот, а потоа предвидуваме излезни вредности line_y, line_y_ridge, line_y_lasso, line_y_elastic и line_y_bayes со користење на моделите за линеарна, риџ, ласо, еластичнa мрежa и баесова риџ регресија, соодветно. Со помош на дијаграм на растурање може да ги прикажеме податочните точки, со plt.plot() линијата која најдобро одговара на секој модел посебно, а со plt.errorbar() лентите за грешката на моделите за линеарна и баесова риџ регресија. Линијата на секој модел се пресметува во опсегот од 0 до максималната вредност на Xi. Лентите за грешки го покажуваат опсегот на варијацијата околу предвидените вредности на y според соодветниот модел.
# 
# Овој график ни овозможува да ги споредиме предвидените вредности на y од различните модели на регресија и да ги видиме нивните разлики и сличности во однос на релацијата меѓу влезните и излезната променлива. Во овој пример, линиите кои најдобро одговараат на податоците не се премногу блиску до податочните точки, што ни остава простор за истражување и подобрување на моделот.
# 
# 

# In[119]:


# пресметка на линијата на секој модел во опсег [0, X.max()]
line_X = np.arange(0, 1.1, 0.1)[:, np.newaxis]
line_y = lr.predict(line_X)
line_y_ridge = ridge.predict(line_X)
line_y_lasso = lasso.predict(line_X)
line_y_elastic = elastic.predict(line_X)
line_y_bayes, Bayessigma = Bayes.predict(line_X, return_std = True) # во баесовиот модел може да се пресмета варијансата за секоја предикција

lw = 2 # параметар за визуелизација (linewidth)
plt.figure() # ги цртаме податочните точки и ги цртаме линиите што ги претставуваат различните модели, ги цртаме и лентите за грешки на моделите за линеарна и баесова риџ регресија
plt.scatter(Xi, y, color='darkgreen', marker='.', s = 300)
plt.plot(line_X, line_y, color='navy', linewidth=lw, label='Линеарна регресија', alpha=0.5)
plt.errorbar(line_X, line_y,yerr=lrsigma)
plt.plot(line_X, line_y_ridge, color='cornflowerblue', linewidth=lw, label='Риџ регресија', alpha=0.5)
plt.plot(line_X, line_y_lasso, color='darkorange', linewidth=lw, label='Ласо регресија', alpha=0.5)
plt.plot(line_X, line_y_elastic, color="red", linewidth=lw, label='Eластичнa мрежa', alpha=0.5)
plt.plot(line_X, line_y_bayes, color="yellow", linewidth=lw, label='Баесова риџ регресија', alpha=0.5)
plt.errorbar(line_X, line_y_bayes, yerr=Bayessigma, alpha = 0.3)

plt.legend(loc='best')
plt.xlabel("Влез")
plt.ylabel("Излез")
plt.show()


# ### Проширување на основниот линеарен модел во квадратен модел, споредба на моделите со приказ на линијата која најдобро одговара на податоците
# Бидејќи податоците не одговараат најдобро на линеарниот модел на регресија, може да направиме проширување на основната функција од линеарен случај во квадратен. Првиот атрибут кој претходно го разгледувавме Xi сега ќе има две карактеристики (едната за $X$, другата за $X^2$). Повторно ќе ги поставиме моделите на регресија (линеарна, риџ, ласо, еластична мрежа и Баесова риџ), но овој пат за да може да ги споредиме да ја увидиме зависноста на квадратниот модел на првиот атрибут Xi (Xt) во однос на излезната променлива y. Со користење на моделите, ги пресметуваме коефициентите за секој модел, кои ги опишуваат зависноста меѓу променливата Xt и излезната променлива y. Отпечатените коефициенти на моделите (по два за секој модел: еден за $X$, а еден за $X^2$) ни даваат информација за влијанието и значењето на Xt во предвидувањето на y. Што поголема вредност има коефициентот, тоа повеќе влијание има Xt врз y. Кај сите модели може да забележиме поголемо влијание кај првиот коефициент (линеарниот) отколку вториот (квадратниот). Ова укажува на тоа дека не постигнуваме некое значајно подобрување со предложеното проширување. Оваа анализа на коефициентите ни помага да го разбереме влијанието на конкретниот проширен атрибут Xt врз излезната променлива y и да ги споредиме перформансите на различните модели на регресија во однос на овој атрибут.
# 
# На сличен начин како и претходно, ги прикажуваме линијиите која најдобро одговараат на претходните модели. Со помош на дијаграм на растурање ги прикажуваме податочните точки, линијата која најдобро одговара на секој модел посебно, и лентите за грешката на моделите за линеарна и баесова риџ регресија. Линијата на секој модел сега не е линеарна, туку со облик на квадратна функција. Лентите за грешки го покажуваат опсегот на варијацијата околу предвидените вредности на y според соодветниот модел.
# 
# Графикот ни овозможува да ги споредиме предвидените вредности на y од различните проширени модели на регресија и да ги видиме нивните разлики и сличности во однос на релацијата меѓу влезните и излезната променлива. Повторно во овој пример, линиите кои најдобро одговараат на податоците не се премногу блиску до податочните точки, и покрај промената во квадратен модел. Од графикот може да забележиме дека проширените квадратни модели на линеарна и риџ регресија привидно најмногу одговараат.
# 

# In[120]:


p = np.array([1, 2]) # за едноставен еднодимензионален случај, можеме да направиме проширување на основната функција вака
Xt = Xi ** p # забележете дека Xt сега има две карактеристики (едната претставува X, а другата претставува X^2)
print("Обликот на X: ", Xi.shape)
print("Обликот на трансформираната матрица Xt", Xt.shape)

print(Xt.shape)
print(y.shape)

# поставете модел на линеарна регресија
lr = LinearRegression() # иницијализација на моделот
lr.fit(Xt, y) # поставете го моделот на податоците
lrsigma = np.sqrt(np.sum((y - lr.predict(Xt))**2) /(N-2)) # пресметка на варијанската на моделот

# поставете модел на риџ регресија
ridge = Ridge(alpha=1)
ridge.fit(Xt, y)
ridgesigma = np.sum((y - ridge.predict(Xt))**2) /(N-2)

# поставете модел на ласо регресија
lasso = Lasso(alpha=0.1)
lasso.fit(Xt, y)
lassosigma = np.sum((y - lasso.predict(Xt))**2) /(N-2)

# поставете модел на регресија на еластична мрежа
elastic = ElasticNet(alpha=0.1, l1_ratio = 0.4)
elastic.fit(Xt, y)
elasticsigma = np.sum((y - elastic.predict(Xt))**2) /(N-2)

# поставете модел на баесовa риџ регресија
Bayes = BayesianRidge()
Bayes.fit(Xt, y)

print("Коефициенти на моделите")
print("Линеарна регресија: ",lr.coef_)
print("Риџ регресија: ", ridge.coef_)
print("Ласо регресија: ", lasso.coef_)
print("Еластична мрежа:", elastic.coef_)
print("Баесова риџ регресија: ", Bayes.coef_)

# пресметка на линијата на секој модел во опсег [0, X1.max()]
line_X0 = np.arange(0, 1.1, 0.1)[:, np.newaxis]
line_X1 = line_X0*line_X0
line_X = np.concatenate((line_X0,line_X1),1)
print(line_X)
line_y = lr.predict(line_X)
line_y_ridge = ridge.predict(line_X)
line_y_lasso = lasso.predict(line_X)
line_y_elastic = elastic.predict(line_X)
line_y_bayes, Bayessigma = Bayes.predict(line_X, return_std = True) # во баесовиот модел може да се пресмета варијансата за секоја предикција

M=line_X1.size
lrsigma_array=np.ones(M)*lrsigma

lw = 2 # параметар за визуелизација (linewidth)
plt.figure() # ги цртаме податочните точки и ги цртаме линиите што ги претставуваат различните модели, ги цртаме и лентите за грешки на моделите за линеарна и баесова риџ регресија
plt.scatter(Xi, y, color='darkgreen', marker='.', s = 300)
plt.plot(line_X0, line_y, color='navy', linewidth=lw, label='Линеарна регресија', alpha=0.5)
plt.errorbar(line_X0, line_y.flatten(), yerr=lrsigma_array)
plt.plot(line_X0, line_y_ridge, color='cornflowerblue', linewidth=lw, label='Риџ регресија', alpha=0.5)
plt.plot(line_X0, line_y_lasso, color='darkorange', linewidth=lw, label='Ласо регресија', alpha=0.5)
plt.plot(line_X0, line_y_elastic, color="red", linewidth=lw, label='Еластична мрежа', alpha=0.5)
plt.plot(line_X0, line_y_bayes, color="yellow", linewidth=lw, label='Баесова риџ регресија', alpha=0.5)
plt.errorbar(line_X0, line_y_bayes, yerr=Bayessigma, alpha = 0.3)

plt.legend(loc='best')
plt.xlabel("Влез")
plt.ylabel("Излез")
plt.show()


# ### Селекција на карактеристики (атрибути)
# Бидејќи поставениот модел на повеќекратна регресија не е многу добар, може да направиме селекција на најдобрите карактеристики (атрибути) од податочното множество. SelectKBest е метод за селекција на карактеристики кој ги оценува карактеристиките според нивната врска со целната променлива. Во овој случај, се користи f_regression како функција за оценка на врската со целната променлива, го поставуваме бројот на најдобри карактеристики на 5 што треба да се изберат.
# fit_transform се користи за тренирање на селектираните атрибути и трансформација на тренирачкото податочно множество, додека transform се користи за трансформација на тест множеството. Ја печатиме маската што означува кои карактеристики се избрани (1) и кои се отфрлени (0), па и новата форма на тренирачкото множество откако селекцијата на карактеристики е извршена.
# 

# In[121]:


selector = SelectKBest(f_regression, k = 5)
X_train_new = selector.fit_transform(X_train, y_train)
X_test_new = selector.transform(X_test)
print(selector.get_support().astype(int))
print(X_train_new.shape)


# Потоа, правиме повеќекратна линеарна регресија врз новите тренирачките податоци X_train-new и соодветните излезни вредности y_train_new. Ги предвидуваме вредностите на излезните податоци y за тест податоците X_test_new. За оценување на фитуваниот модел на линерна регресија пресметуваме различни метрики: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Може да забележиме дека и со селекцијата на атрибути резултатите не се менуваат многу.

# In[122]:


reg_new = LinearRegression()
reg_new.fit(X_train_new, y_train)
y_pred_new = reg_new.predict(X_test_new)
print("Средна квадратна грешка (ново):", mean_squared_error(y_test, y_pred_new))
print("Средна апсолутна грешка (ново):", mean_absolute_error(y_test, y_pred_new))
print("Корен на средна квадратна грешка (ново):", np.sqrt(mean_squared_error(y_test, y_pred_new)))
print("R-квадрат (ново): ", r2_score(y_test,y_pred_new)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test, y_pred_new)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.title("Линеарна регресија на California housing податочно множество (Top 5 атрибути)")
plt.show()


# ## Пример за линеарна регресија со податочното множество Oring
# 
# Податочното множество Oring се состои од податоци за критичните затворачи наречени О-прстени за 23 шатл-мисии, каде што редоследот на мисијата се заснова на температурата во моментот на лансирањето. Оштетувањето на овие О-прстени за време на лансирањето на шатлот може да биде поврзано со температурата на околината за време на лансирањето.
# За секој од податоците од 23те набљудувања се чуваат следните 4 променливи, од кои 3 влезни: температура, притисок и број на мисијата на шатлот како и излезната променлива: број на оштетени О-прстени.
# 
# Наша задача ќе биде да ја предвидиме целната променлива од влезните променливи и да поставиме модел кој најдобро одговара на повеќекратна регресија која ќе ја предвиди оваа излезна променлива со податоците од влезните 3 атрибути. Ќе ја испитаме зависноста на секој атрибут со излезната променлива, за да може да се најдат атрибутите кои најмногу може да придонесат за подобар регресионен модел.
# 
# Прво, ќе покажеме како може да се вчита податочното множество, а потоа како може да се направи основна анализа и визуелизација на податоците, за да на крај ја прикажеме можноста за вклопување на различни модели на регресија.
# 
# #### Вчитување на потребните библиотеки
# Најпрво треба да ги вчитаме библиотеките кои се потребни за извршување на кодот во оваа задача.

# In[123]:


import matplotlib.pyplot as plt #Библиотека за визуелизација
import seaborn as sns
import numpy as np #Библиотека за нумерички анализи
import pandas as pd
from scipy import stats #Библиотека за статистички анализи
from sklearn.model_selection import train_test_split
from sklearn import datasets #Библиотека за податочни множества за машинско учење
from sklearn.preprocessing import MinMaxScaler #Min max Scaler за нормализација на податоци во опсег [0,1]
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge #Вчитување на модели за линерна регресија
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score #За евалуација на модели


# ### Вчитување на податоците
# Податочното множество може да го вчитаме со функциски повик за читање на .csv датотеки како "pandas" податочна рамка. Како втора опција која ќе ја разгледаме за вчитување на податочното множество е преку повици за читање од numpy библиотеката.

# In[124]:


df_data = pd.read_csv("datasets/Oring.csv") # Вчитување на податочното множество
column_names = np.loadtxt(open("datasets/Oring.csv", "rb"), delimiter=",", max_rows=1, dtype = str)
data = np.loadtxt(open("datasets/Oring.csv", "rb"), delimiter=",", skiprows=1, dtype = str)
data = data.astype(float)


# ### Визуелизација на податоците
# Ќе ги прикажеме првите 5 податоци од множеството.

# In[125]:


df_data.head(5)


# Ја иницијализираме влезната променлива X, каде што секој ред одговара на податочна точка, а секоја колона одговара на карактеристика (атрибут), a и излезната променлива y каде што i-тата вредност е излезната цена на i-тиот вектор на атрибут (ред) во X. Потоа, може да ги испечатиме имињата на атрибутите, бројот на атрибути (3), бројот на излезни променливи (1), различните вредности на излезната променлива и бројот на инстанци (23) во множеството.
# 
# 

# In[126]:


X = data[:,1:].astype(float) # другите колони ги претставуваат влезните податоци
y = data[:,0].astype(float)  # првата колона ја претставува излезната променлива
N = X.shape[0]
number_of_features = X.shape[1]
feature_names = column_names[1:]

print(X.shape)
print(y.shape)
print("Карактеристики (Атрибути): ", feature_names)
print("Број на карактеристики (атрибути):", X.shape[1]) # или number_of_features
print("Број на излезни класи: ", len(np.unique(y)))
print("Број на инстанци: ", X.shape[0]) # или N


# Распределбата на вредностите за атрибутите може да ја разгледаме со користење на хистограм. Тука имаме 3 атрибути, па затоа имаме 3 хистограми посебно. Истото го правиме за да прикажеме хистограм на вредностите на излезната променлива. Може да забележиме дека обликот на хистограмите на влезните атрибути, а и на излезната променлива, е различен.
# 

# In[127]:


fig = plt.figure(figsize=(8, 8))
fig.subplots(number_of_features//2+1, ncols=2)
for feat_i in range(number_of_features): # хистограми за секој атрибут
    ax = plt.subplot(number_of_features//2+1,2, feat_i+1)
    plt.title(feature_names[feat_i])
    sns.histplot(X[:,feat_i], color = "navy")
ax = plt.subplot(number_of_features//2+1,2, feat_i+2)
plt.title("Излез: " + column_names[0])
sns.histplot(y, color = "darkorange") # хистограм за излезната променлива
plt.show()


# ### Визуелизација на зависности
# Може да ја прикажеме зависноста на секој атрибут со секој друг атрибут со визуелизација на атрибутите (карактеристиките) преку дијаграми на растурање. Прикажуваме 3x3=9 вакви дијаграми, по еден за секоја комбинација на два атрибути. Оваа визуелизација ни помага да го истражиме односот помеѓу различните атрибути во податочното множество и да забележиме можни зависности или корелации помеѓу нив. На прв поглед, дијаграмите се многу "растурени", но визуелно може да се забележат некоја мала линеарна зависност меѓу атрибутите 1 и 2.

# In[128]:


fig = plt.figure(figsize=(10, 10))
fig.subplots(nrows=number_of_features, ncols=number_of_features)
for feat_i in range(number_of_features): # за сите парови атрибути (3x3 во овој случај)
    for feat_j in range(number_of_features):
        ax = plt.subplot(number_of_features,number_of_features,number_of_features*feat_i + feat_j+1)
        if feat_i==feat_j:
          plt.hist(X[:, feat_i], alpha=.6, color="orange")
          plt.title("Атрибут "+str(feat_i)+" (хистограм)")
        else:
          plt.scatter(X[:,feat_i], X[:,feat_j],  alpha=.8, color="darkorange")
          plt.title("Атрибут "+str(feat_i)+" x Атрибут "+str(feat_j)) # Атрибут i x Атрибут ј
        plt.yticks(())
plt.show()


# Може да ја прикажеме зависноста на секој атрибут и со излезната променлива преку дијаграми на растурање. Прикажуваме 3 вакви дијаграми, по еден за секој атрибут со излезната променлива.
# 

# In[129]:


fig = plt.figure(figsize=(10, 3))
fig.subplots(ncols=number_of_features)
for feat_i in range(number_of_features):
    ax = plt.subplot(1 ,number_of_features, feat_i+1)
    plt.scatter( X[:,feat_i], y,  alpha=.8, color="darkorange")
    plt.xticks(())
    plt.yticks(())
    plt.title("Атрибут "+str(feat_i)+" x Излез")
plt.show()


# Со контурниот график на густина ја претставуваме распределбата на податоците и можни области на концентрација на зависностите помеѓу атрибутите и излезните вредности.

# In[130]:


for feat_i in range(number_of_features):
    sns.jointplot(x=X[:,feat_i], y=y, kind="kde", space=0, color="g", height = 3)
plt.show()


# Ззависностите може да ги изразиме квантитативно, со помош на Пирсоновиот коефициент на корелација. Затоа, пресметуваме и визуелизираме корелациска матрица во која се претставени Пирсоновите коефициенти на корелација за сите парови атрибути, како и за секој атрибут со целната променлива, во последната редица и колона на матрицата. Од овде, може да се види зависноста меѓу атрибут 0 и излезната променлива.

# In[131]:


correlation_matrix = np.ones((number_of_features +1,number_of_features +1))
# прво ги пресметуваме пирсоновите корелациски коефициенти за сите парови атрибути
for i in range(number_of_features):
    measure = X[:,i]
    for j in range(number_of_features):
        measure2 = X[:,j]
        corr, _ = stats.pearsonr(measure, measure2)
        correlation_matrix[i][j] = corr
# потоа ги пресметуваме пирсоновите корелациски коефициенти за секој атрибут со целната променлива
for i in range(number_of_features):
    j = number_of_features
    measure = X[:,i]
    measure2 = y
    corr, _ = stats.pearsonr(measure, measure2)
    correlation_matrix[i][j] = corr
    correlation_matrix[j][i] = corr
plt.figure()
plt.imshow(correlation_matrix, cmap = "inferno", vmin=-1, vmax=1) #матрицата може да се исцрта со imshow
plt.colorbar()
plt.show()


# ### Нормализација на податоците
# Ќе направиме мин-макс нормализација на податоците со MinMaxScaler() објект, како и во претходниот пример за повеќекратна линеарна регресија. Потоа, ги делиме нормализираните податоци во две групи - тренирачка група и тест група. X_train содржи тренирачки влезни податоци (обележени со X) кои се користат за обучување на моделот, X_test содржи тест влезни податоци (обележени со X) кои се користат за проверка на моделот, y_train содржи тренирачки излезни податоци (обележени со y) кои се користат за обучување на моделот, a y_test содржи тест излезни податоци (обележени со y) кои се користат за проверка на моделот. Го поставуваме односот помеѓу тренирачките и тест податоци на 30% за тест.
# 

# In[132]:


scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
#Забележете дека бидејќи ова е малo податочно множество, не треба да го користиме стандардниот train_test_split
#за да ја процениме точноста на множеството податоци, туку треба да користиме вкрстена валидација.
#Меѓутоа, за едноставност, го правиме тоа во овој пример.


# ### Линеарна регресија и пресметка на грешки
# Потоа, правиме повеќекратна линеарна регресија како во претходните примери врз тренирачките податоци X_train и соодветните излезни вредности y_train. Се пресметуваат коефициентите (W) и пресекот (W0) на регресијата. Може да забележиме дека има 3 коефициенти, по еден за секој атрибут. Со помош на reg.predict(X_test), се предвидуваат вредностите на излезните податоци y за тест податоците X_test. За оценување на фитуваниот модел на линерна регресија пресметуваме различни метрики: средна квадратна грешка, средна апсолутна грешка, корен на средна квадратна грешка и R-квадрат. Може да забележиме дека грешките се мали, а R-квадрат вредноста е околу -0.18, што укажува на мала негативна линеарна врска и тоа дека моделот не е многу добар во објаснувањето на варијациите во податоците.
# 

# In[133]:


reg = LinearRegression().fit(X_train, y_train)
print("Коефициенти W: ", reg.coef_)
print("Пресек W0: ", reg.intercept_)

y_pred = reg.predict(X_test) # предвиди ги вредностите на тест податоците
print("Средна квадратна грешка:", mean_squared_error(y_test, y_pred))
print("Средна апсолутна грешка:", mean_absolute_error(y_test, y_pred))
print("Корен на средна квадратна грешка:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R-квадрат: ", r2_score(y_test,y_pred)) # пресметај ја R - квадрат вредноста

# визуелизација на предвидените вредности од моделот
plt.figure(figsize=(4, 4))
plt.scatter(y_test,y_pred)
plt.xlabel("Вистински вредности")
plt.ylabel("Предвидени вредности")
plt.show()


# Може да ги претставиме и добиените тежини на коефициентите за секој атрибут.
# 

# In[134]:


plt.figure(figsize=(4, 4))
plt.plot(np.arange(len(reg.coef_)), reg.coef_, ".",  markersize=15, color = "darkgreen")
plt.xlabel("Број на атрибут")
plt.ylabel("Коефициенти w")
plt.title("Тежина на коефициенти")
plt.show()


# **Задача 1.** Поставете ги сите останати регресиони модели кои ги споменавме: риџ, ласо, еластична мрежа и баесова риџ регресија на податочното множество О-ринг. Направете споредба на резултатите, кој модел најдобро одговара на податоците?
# 
# 

# ### Споредба на моделите со приказ на линијата која најдобро одговара на податоците
# Ќе ги поставиме моделите на регресија (линеарна, риџ, ласо, еластична мрежа и Баесова риџ), за да може да ги споредиме и тоа само да ја увидиме зависноста на првиот атрибут (атрибут 0) X во однос на излезната променлива y. Со користење на моделите, ги пресметуваме коефициентите за секој модел, кои ги опишуваат зависноста меѓу променливата X и излезната променлива y. Отпечатените коефициенти на моделите ни даваат информација за влијанието и значењето на X во предвидувањето на y.
# 

# In[135]:


column_names = np.loadtxt(open("datasets/Oring.csv", "rb"), delimiter=",", max_rows=1, dtype = str)
data = np.loadtxt(open("datasets/Oring.csv", "rb"), delimiter=",", skiprows=1, dtype = str)
data = data.astype(float)
X = data[:,[1]].astype(float)  # атрибут 0
y = data[:,0].astype(float)
number_of_features = X.shape[1]
feature_names = column_names[1]
N = X.shape[0]

# поставете модел на линеарна регресија
lr = LinearRegression() # иницијализација на моделот
lr.fit(X, y) # поставете го моделот на податоците
lrsigma = np.sum((y - lr.predict(X))**2) /(N-2) # пресметка на варијансата на моделот

# поставете модел на риџ регресија
ridge = Ridge(alpha=1)
ridge.fit(X, y)
ridgesigma = np.sum((y - ridge.predict(X))**2) /(N-2)

# поставете модел на ласо регресија
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
lassosigma = np.sum((y - lasso.predict(X))**2) /(N-2)

# поставете модел на регресија на еластична мрежа
elastic = ElasticNet(alpha=0.1, l1_ratio = 0.4)
elastic.fit(X, y)
elasticsigma = np.sum((y - elastic.predict(X))**2) /(N-2)

# поставете модел на баесова риџ регресија
Bayes = BayesianRidge()
Bayes.fit(X, y)

print("Коефициенти на моделите")
print("Линеарна регресија: ",lr.coef_)
print("Риџ регресија: ", ridge.coef_)
print("Ласо регресија: ", lasso.coef_)
print("Еластична мрежа:", elastic.coef_)
print("Баесова риџ регресија: ", Bayes.coef_)


# За да ја прикажеме линијата која најдобро одговара на претходните модели, поставуваме линија line_X за претставување на линијата на моделот, а потоа предвидуваме излезни вредности line_y, line_y_ridge, line_y_lasso, line_y_elastic и line_y_bayes со користење на моделите за линеарна, риџ, ласо, еластичн мрежа и баесова риџ регресија, соодветно. Со помош на еден дијаграм може да ги
# прикажеме податочните точки, линијата која најдобро одговара на секој модел посебно, и лентите за грешката на моделите за линеарна и баесова риџ регресија. Линијата на секој модел се пресметува во опсегот од 20 до максималната вредност на X. Лентите за грешки го покажуваат опсегот на варијацијата околу предвидените вредности на y според соодветниот модел.
# 
# Во овој пример, може да забележиме дека линиите кои најдобро одговараат на податоците на сите модели не се многу блиску до податочните точки, дијаграмот е доста растурен, но има многу мала линеарна зависност на избраниот атрибут со излезната променлива.
# 

# In[136]:


# пресметка на линијата на секој модел во опсег [20, X.max()]
line_X = np.arange(20, X.max())[:, np.newaxis]
line_y = lr.predict(line_X)
line_y_ridge = ridge.predict(line_X)
line_y_lasso = lasso.predict(line_X)
line_y_elastic = elastic.predict(line_X)
line_y_bayes, Bayessigma = Bayes.predict(line_X, return_std = True) # во баесовиот модел може да се пресмета варијансата за секоја предикција

lw = 2 # параметар за визуелизација (linewidth)
plt.figure() # ги цртаме податочните точки и ги цртаме линиите што ги претставуваат различните модели, ги цртаме и лентите за грешки на моделите за линеарна и баесова Ridge регресија
plt.scatter(X, y, color='darkgreen', marker='.', s = 300)
plt.plot(line_X, line_y, color='navy', linewidth=lw, label='Линеарна регресија', alpha=0.5)
plt.errorbar(line_X, line_y,yerr=lrsigma)
plt.plot(line_X, line_y_ridge, color='cornflowerblue', linewidth=lw, label='Риџ регресија', alpha=0.5)
plt.plot(line_X, line_y_lasso, color='darkorange', linewidth=lw, label='Ласо регресија', alpha=0.5)
plt.plot(line_X, line_y_elastic, color="red", linewidth=lw, label='Еластична мрежа', alpha=0.5)
plt.plot(line_X, line_y_bayes, color="yellow", linewidth=lw, label='Баесова риџ регресија', alpha=0.5)
plt.errorbar(line_X, line_y_bayes, yerr=Bayessigma, alpha = 0.3)

plt.legend(loc='best')
plt.xlabel("Влез")
plt.ylabel("Излез")
plt.show()


# ### Проширување на основниот линеарен модел во квадратен модел, споредба на моделите со приказ на линијата која најдобро одговара на податоците
# Може да направиме проширување на основната функција од линеарен случај во квадратен. Првиот атрибут кој претходно го разгледувавме X (атрибут 0) сега ќе има две карактеристики (едната за $X$, другата за $X^2$). Повторно ќе ги поставиме моделите на регресија (линеарна, риџ, ласо, еластична мрежа и Баесова риџ), но овој пат за да може да ги споредиме да ја увидиме зависноста на квадратниот модел на првиот атрибут X (Xt) во однос на излезната променлива y. Со користење на моделите, ги пресметуваме коефициентите за секој модел, кои ги опишуваат зависноста меѓу променливата Xt и излезната променлива y. Отпечатените коефициенти на моделите (по два за секој модел: еден за $X$, а еден за $X^2$) ни даваат информација за влијанието и значењето на Xt во предвидувањето на y. Што поголема вредност има коефициентот, тоа повеќе влијание има Xt врз y. Кај сите модели може да забележиме различно влијание кај првиот коефициент (линеарниот) со вториот (квадратниот). Оваа анализа на коефициентите ни помага да го разбереме влијанието на конкретниот проширен атрибут Xt врз излезната променлива y и да ги споредиме перформансите на различните модели на регресија во однос на овој атрибут.
# 
# На сличен начин како и претходно, ги прикажуваме линиите која најдобро одговараат на претходните модели. Со помош на дијаграм на растурање ги прикажуваме податочните точки, линијата која најдобро одговара на секој модел посебно, и лентите за грешката на моделите за линеарна и баесова риџ регресија. Линијата на секој модел сега не е линеарна, туку со облик на квадратна функција. Лентите за грешки го покажуваат опсегот на варијацијата околу предвидените вредности на y според соодветниот модел.
# 

# In[137]:


p = np.array([1, 2]) # за едноставен еднодимензионален случај, можеме да направиме проширување на основната функција вака
Xt = X ** p # забележете дека Xt сега има две карактеристики (едната претставува X, а другата претставува X^2)
print("Обликот на X: ", X.shape)
print("Обликот на трансформираната матрица Xt", Xt.shape)

# поставете модел на линеарна регресија
lr = LinearRegression() # иницијализација на моделот
lr.fit(Xt, y) # поставете го моделот на податоците
lrsigma = np.sqrt(np.sum((y - lr.predict(Xt))**2) /(N-2)) # пресметка на варијанската на моделот

# поставете модел на риџ регресија
ridge = Ridge(alpha=1)
ridge.fit(Xt, y)
ridgesigma = np.sum((y - ridge.predict(Xt))**2) /(N-2)

# поставете модел на ласо регресија
lasso = Lasso(alpha=0.1)
lasso.fit(Xt, y)
lassosigma = np.sum((y - lasso.predict(Xt))**2) /(N-2)

# поставете модел на регресија на еластична мрежа
elastic = ElasticNet(alpha=0.1, l1_ratio = 0.4)
elastic.fit(Xt, y)
elasticsigma = np.sum((y - elastic.predict(Xt))**2) /(N-2)

# поставете модел на баесовa риџ регресија
Bayes = BayesianRidge()
Bayes.fit(Xt, y)

print("Коефициенти на моделите")
print("Линеарна регресија: ",lr.coef_)
print("Риџ регресија: ", ridge.coef_)
print("Ласо регресија: ", lasso.coef_)
print("Еластична мрежа:", elastic.coef_)
print("Баесова риџ регресија: ", Bayes.coef_)

# пресметка на линијата на секој модел во опсег [0, X1.max()]
line_X = np.arange(20, X.max())[:, np.newaxis]
line_y = lr.predict(line_X ** p)
line_y_ridge = ridge.predict(line_X ** p)
line_y_lasso = lasso.predict(line_X ** p)
line_y_elastic = elastic.predict(line_X ** p)
line_y_bayes, Bayessigma = Bayes.predict(line_X**p, return_std = True) # во баесовиот модел може да се пресмета варијансата за секоја предикција

lw = 2 # параметар за визуелизација (linewidth)
plt.figure() # ги цртаме податочните точки и ги цртаме линиите што ги претставуваат различните модели, ги цртаме и лентите за грешки на моделите за линеарна и баесова риџ регресија
plt.scatter(X, y, color='darkgreen', marker='.', s = 300)
plt.plot(line_X, line_y, color='navy', linewidth=lw, label='Линеарна регресија', alpha=0.5)
plt.errorbar(line_X, line_y.flatten(), yerr=lrsigma)
plt.plot(line_X, line_y_ridge, color='cornflowerblue', linewidth=lw, label='Риџ регресија', alpha=0.5)
plt.plot(line_X, line_y_lasso, color='darkorange', linewidth=lw, label='Ласо регресија', alpha=0.5)
plt.plot(line_X, line_y_elastic, color="red", linewidth=lw, label='Еластична мрежа', alpha=0.5)
plt.plot(line_X, line_y_bayes, color="yellow", linewidth=lw, label='Баесова риџ регресија', alpha=0.5)
plt.errorbar(line_X, line_y_bayes, yerr=Bayessigma, alpha = 0.3)

plt.legend(loc='best')
plt.xlabel("Влез")
plt.ylabel("Излез")
plt.show()


# ### Полиномно проширување на основниот линеарен модел во кубен модел, споредба на моделите со приказ на линијата која најдобро одговара на податоците
# Може повторно да направиме уште едно полиномно проширување на основната функција од квадратен случај во кубен. Првиот атрибут кој претходно го разгледувавме X (атрибут 0) сега ќе има три карактеристики (едната за $X$, втората за $X^2$ и третата за $X^3$). Повторно ќе ги поставиме моделите на регресија (линеарна, риџ, ласо, еластична мрежа и Баесова риџ), но овој пат за да може да ги споредиме да ја увидиме зависноста на кубниот модел на првиот атрибут X (Xt) во однос на излезната променлива y. Со користење на моделите, ги пресметуваме коефициентите за секој модел, кои ги опишуваат зависноста меѓу променливата Xt и излезната променлива y. Отпечатените коефициенти на моделите (по три за секој модел: еден за $X$, втор за $X^2$ и трет за $X^3$) ни даваат информација за влијанието и значењето на Xt во предвидувањето на y. Што поголема вредност има коефициентот, тоа повеќе влијание има Xt врз y. Кај сите модели може да забележиме различно влијание кај првиот коефициент (линеарниот) со вториот (квадратниот) и третиот (кубниот). Оваа анализа на коефициентите ни помага да го разбереме влијанието на конкретниот проширен атрибут Xt врз излезната променлива y и да ги споредиме перформансите на различните модели на регресија во однос на овој атрибут.
# 
# На сличен начин како и претходно, ги прикажуваме линиите која најдобро одговараат на претходните модели. Со помош на дијаграм на растурање ги прикажуваме податочните точки, линијата која најдобро одговара на секој модел посебно, и лентите за грешката на моделите за линеарна и баесова риџ регресија. Линијата на секој модел сега не е линеарна, туку со облик на кубна функција. Лентите за грешки го покажуваат опсегот на варијацијата околу предвидените вредности на y според соодветниот модел. Визуелно може да забележиме дека овој модел подобро одговара на податоците. Сепак, поради само 3 можни вредности на излезната променлива, регресионите модели не се најсоодветни во овој случај.
# 

# In[138]:


from sklearn import linear_model
p = np.array([1, 2, 3])
Xt = X ** p # забележете дека Xt сега има 3 атрибути (еден за X, еден за X^2 и еден за X^3)
print("Обликот на X: ", X.shape)
print("Обликот на трансформираната матрица Xt", Xt.shape)

# поставете модел на линеарна регресија
lr = LinearRegression() # иницијализација на моделот
lr.fit(Xt, y) # поставете го моделот на податоците
lrsigma = np.sqrt(np.sum((y - lr.predict(Xt))**2) /(N-2)) # пресметка на варијанската на моделот

# поставете модел на риџ регресија
ridge = Ridge(alpha=1)
ridge.fit(Xt, y)
ridgesigma = np.sum((y - ridge.predict(Xt))**2) /(N-2)

# поставете модел на ласо регресија
lasso = Lasso(alpha=0.1)
lasso.fit(Xt, y)
lassosigma = np.sum((y - lasso.predict(Xt))**2) /(N-2)

# поставете модел на регресија на еластична мрежа
elastic = ElasticNet(alpha=0.1, l1_ratio = 0.4)
elastic.fit(Xt, y)
elasticsigma = np.sum((y - elastic.predict(Xt))**2) /(N-2)

# поставете модел на баесовa риџ регресија
Bayes = BayesianRidge()
Bayes.fit(Xt, y)

print("Коефициенти на моделите")
print("Линеарна регресија: ",lr.coef_)
print("Риџ регресија: ", ridge.coef_)
print("Ласо регресија: ", lasso.coef_)
print("Еластична мрежа:", elastic.coef_)
print("Баесова риџ регресија: ", Bayes.coef_)

# пресметка на линијата на секој модел во опсег [20, X.max()]
line_X = np.arange(20, X.max())[:, np.newaxis]
line_y = lr.predict(line_X ** p)
line_y_ridge = ridge.predict(line_X ** p)
line_y_lasso = lasso.predict(line_X ** p)
line_y_elastic = elastic.predict(line_X ** p)
line_y_bayes, Bayessigma = Bayes.predict(line_X**p, return_std = True) # во баесовиот модел може да се пресмета варијансата за секоја предикција

lw = 2 # параметар за визуелизација (linewidth)
plt.figure() # ги цртаме податочните точки и ги цртаме линиите што ги претставуваат различните модели, ги цртаме и лентите за грешки на моделите за линеарна и баесова риџ регресија
plt.scatter(X, y, color='darkgreen', marker='.', s = 300)
plt.plot(line_X, line_y, color='navy', linewidth=lw, label='Линеарна регресија', alpha=0.5)
plt.errorbar(line_X, line_y.flatten(), yerr=lrsigma)
plt.plot(line_X, line_y_ridge, color='cornflowerblue', linewidth=lw, label='Риџ регресија', alpha=0.5)
plt.plot(line_X, line_y_lasso, color='darkorange', linewidth=lw, label='Ласо регресија', alpha=0.5)
plt.plot(line_X, line_y_elastic, color="red", linewidth=lw, label='Еластична мрежа', alpha=0.5)
plt.plot(line_X, line_y_bayes, color="yellow", linewidth=lw, label='Баесова риџ регресија', alpha=0.5)
plt.errorbar(line_X, line_y_bayes, yerr=Bayessigma, alpha = 0.3)

plt.legend(loc='best')
plt.xlabel("Влез")
plt.ylabel("Излез")
plt.show()


# **Задача 2.** Поставете ги сите регресиони модели кои ги споменавме: линеарна, риџ, ласо, еластична мрежа и Баесова риџ регресија на податочното множество diabetes од библиотеката sklearn.datasets (множеството се вчитува со функцискиот повик load_diabetes). Направете ги сите чекори за визуелизација на множеството, зависностите на атрибутите меѓусебно, а и со излезните променливи. Поставете и проширен квадратен и кубен регресионен модел на податоците, а и прикажете ги линиите кои најдобро одговараат на секој модел. Направете споредба на резултатите, кој модел најдобро одговара на податоците?
# 
# **Задача 3.** Поставете ги сите регресиони модели кои ги споменавме: линеарна, риџ, ласо, еластична мрежа и
# Баесова риџ регресија на податочното множество linerud од библиотеката sklearn.datasets (множеството се вчитува со функцискиот повик load_linerud). Направете ги сите чекори за визуелизација на множеството, зависностите на атрибутите меѓусебно, а и со излезните променливи. Поставете и проширен квадратен и кубен регресионен модел на податоците, а и прикажете ги линиите кои најдобро одговараат на секој модел. Направете споредба на резултатите, кој модел најдобро одговара на податоците?
# 
# **Задача 4.** Поставете ги сите регресиони модели кои ги споменавме: линеарна, риџ, ласо, еластична мрежа и Баесова риџ регресија на податочното множество breast cancer од библиотеката sklearn.datasets (множеството се вчитува со функцискиот повик load_breast_cancer). Направете ги сите чекори за визуелизација на множеството, зависностите на атрибутите меѓусебно, а и со излезните променливи. Поставете и проширен квадратен и кубен регресионен модел на податоците, а и прикажете ги линиите кои најдобро одговараат на секој модел. Направете споредба на резултатите, кој модел најдобро одговара на податоците? Дали моделите на линеарна регресија се соодветни за ова податочно множество?
# 
```
